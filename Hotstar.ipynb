```python
# Convert imdb_rating column to numeric, handling any errors due to non-numeric values
df['imdb_rating'] = pd.to_numeric(df['imdb_rating'], errors='coerce')

# Display the top 10 highest-rated shows/movies
top_rated = df[['title', 'type', 'imdb_rating']].sort_values(by='imdb_rating', ascending=False).head(10)
print(top_rated)
```

```python
# Filter movies only and clean up the runtime column
movies = df[df['type'] == 'movie']
movies['runtime'] = movies['runtime'].str.replace(' min', '').astype(float)

# Calculate and display the average runtime
average_runtime = movies['runtime'].mean()
print(f"Average runtime for movies: {average_runtime} minutes")
```

```python
import matplotlib.pyplot as plt

# Explode the genre column into multiple rows and count occurrences of each genre
genre_counts = df['genre'].str.split(', ').explode().value_counts()

# Plot the genre distribution
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
